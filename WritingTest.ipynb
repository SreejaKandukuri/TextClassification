```python
#KNN
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load the dataset
digits = load_digits()

# Step 2: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)

# Step 3: Feature Scaling
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Train the KNN Model
knn = KNeighborsClassifier(n_neighbors=5)  # Initialize KNN with k=5
knn.fit(X_train_scaled, y_train)

# Step 5: Evaluate the Model
y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Print classification report
print(" KNN Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
#DISPLAYING THE VALUES IN A TABLE - KNN
from sklearn.metrics import classification_report
from tabulate import tabulate


# Generate classification report
report = classification_report(y_test, y_pred, output_dict=True)

# Initialize an empty list to store the table data
table = []

# Loop through each class in the report
for label in report.keys():
    # If the label is not 'accuracy' (since it's not related to a specific class)
    if label != 'accuracy':
        # Extract precision, recall, f1-score, and support for the label
        precision = round(report[label]['precision'], 4)
        recall = round(report[label]['recall'], 4)
        f1_score = round(report[label]['f1-score'], 4)
        support = report[label]['support']
        
        # Append the values to the table list
        table.append([label, precision, recall, f1_score, support])

# Add macro and weighted averages to the table with rounded values
macro_avg_precision = round(report['macro avg']['precision'], 4)
macro_avg_recall = round(report['macro avg']['recall'], 4)
macro_avg_f1_score = round(report['macro avg']['f1-score'], 4)
macro_avg_support = report['macro avg']['support']
table.append(['macro avg', macro_avg_precision, macro_avg_recall, macro_avg_f1_score, macro_avg_support])

weighted_avg_precision = round(report['weighted avg']['precision'], 4)
weighted_avg_recall = round(report['weighted avg']['recall'], 4)
weighted_avg_f1_score = round(report['weighted avg']['f1-score'], 4)
weighted_avg_support = report['weighted avg']['support']
table.append(['weighted avg', weighted_avg_precision, weighted_avg_recall, weighted_avg_f1_score, weighted_avg_support])

# Add accuracy to the table with rounded value
accuracy = round(report['accuracy'], 4)
table.append(['Accuracy', '', '', '', accuracy])

# Display table
print(tabulate(table, headers=['Label', 'Precision', 'Recall', 'F1-Score', 'Support']))
```
